```python

# Transpose the dataframe to get time series for each Component
df_melted = df.melt(id_vars=["Type", "Component"], var_name="Date", value_name="Value")
df_melted["Date"] = pd.to_datetime(df_melted["Date"])

# Pivot for visual correlation
pivot_df = df_melted.pivot(index="Date", columns="Component", values="Value")

# Plot heatmap of correlation
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Financial Components")
plt.tight_layout()
plt.show()
```

```python
# Since sns.pairplot works best with numerical values in wide format, use a subset of relevant components
selected_components = ["Salary", "Freelancig", "Mutual funds", "Fixed Deposit", "Emergency Fund"]
numeric_df = pivot_df[selected_components].dropna()

# Pairplot to see relationships between selected financial components
sns.pairplot(numeric_df)
plt.suptitle("Pairplot of Selected Financial Components", y=1.02)
plt.show()
```

```python
# Set the plot style
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))

# Plot each component as a scatterplot with a trend line
for col in selected_components:
    plt.figure(figsize=(12, 4))
    sns.scatterplot(data=pivot_df, x=pivot_df.index, y=col, s=60, color="blue", label=col)
    sns.lineplot(data=pivot_df, x=pivot_df.index, y=col, color="red", label=f"Trend: {col}")
    plt.title(f"Trend Over Time: {col}", fontsize=14, fontweight='bold')
    plt.xlabel("Date")
    plt.ylabel("Amount")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# Define your selected financial components
selected_components = ["Salary", "Freelancig", "Mutual funds", "Fixed Deposit", "Emergency Fund"]

# Create histograms with KDE plots
plt.figure(figsize=(18, 12))
for i, col in enumerate(selected_components, 1):
    plt.subplot(3, 2, i)
    sns.histplot(data=numeric_df, x=col, kde=True, color="teal", edgecolor="black")
    plt.title(f"{col} Distribution", fontsize=14)
    plt.xlabel("Value")
    plt.ylabel("Frequency")

plt.suptitle("Histograms of Financial Components with KDE Curves", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
